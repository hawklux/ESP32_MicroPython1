### 1. 라디안-각도 이해

```python
n = 3
to_degree = 180 / np.pi # 라디안 -> 각도
to_radian = np.pi / 180 # 각도 -> 라디안

gap_theta = 2 * np.pi / n # 2π를 3등분한 라디안 값 간격 (120도의 라이안값)
```

```python
gap_theta  # 120도의 라디안 값
```

```python
gap_theta * to_degree
```

```python
theta = np.arange(0, 2 * np.pi, gap_theta) # 0부터 2π까지 gap_theta 간격으로 나열
# theta = [gap_theta * i for i in range(n)] # 위와 동일 람다식

theta  # 0, 120, 240도에 해당하는 라디안 값
```

```python
theta = [gap_theta * i for i in range(n)] # 위와 동일 람다식
theta
```

```python
[to_degree * t for t in theta] # 라디안 -> 각도 변환(0, 120, 240도)
```

```python
r = 3 # 원의 반지름
x = [r * np.cos(th) for th in theta] # x좌표
y = [r * np.sin(th) for th in theta] # y좌표
```

### 2. 좌표 리턴 함수 
* n개 만큼 각도 나눠서 
* 좌표(x, y, theta) 도출하고 리턴하기

```python
def get_position(r, n):
    theta_gap = 2 * np.pi / n # 2π를 n등분한 라디안 값 간격
    thetas = [theta_gap * i for i in range(n)] # 0부터 2π까지 gap_theta 간격으로 나열
    x = [r * np.cos(th) for th in thetas] # x좌표
    y = [r * np.sin(th) for th in thetas] # y좌표
    return x, y, thetas
```

```python
get_position(1, 3) # 반지름 1, 4등분 원 ([x 4개], [y 4개], [라디안 4개])
```

### 3. 좌표에 Position 그리기

```python
def drawer(r,n):
    x, y, thetas = get_position(r, n)
    plt.scatter(x, y) # x, y 좌표에 점 찍기
    plt.axis('equal')
    plt.grid() # 격자 표시
    plt.show() # 그래프 표시
```

```python
drawer(1, 10)
```
